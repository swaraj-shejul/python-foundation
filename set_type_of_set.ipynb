# Type of set

# 1) Joined Set
        - intesection (comon) values are available in both the set

```python
A = {2,3,4}
B = {3,4,5}
print(A.isdisjoint(B))           # A and B are not disjoint because they have common values
print(A.intersection(B))          # A and B have common values 3 and 4
```

# 2) Disjoint Set
    - No intersection between sets 

```python
A = {2,3,4}
B = {5,6,7}
A.isdisjoint(B)           # A and B are disjoint because they have no common values
```

# 3) Subset and Superset
    - if all the values of set A is present in the set B then A is called as a subset of B
    - if B is contaning all the values of A then B is called as superset of A

```python
A = {2,3,4}
B = {2,3,4,5}

A.issubset(B)        # A is subset of B because all values of A are in B
```

```python
A = {1,2,3}
B = {1,2,3,4,5}
B.issuperset(A)      # B is superset of A because all values of A are in B
```
