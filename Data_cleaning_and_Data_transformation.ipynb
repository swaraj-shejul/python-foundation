# Data cleaning for structured data

- for every data cleaning step we have 3 options, out of which we have select any 1

1) Replace
   - replace with original data by contacting backend team (best solution)
   - replace with statistically
     1) numerical column : replace with mean
     2) text/categorical column : replace with mode (most repeated) 
2) Remove
   wrong data less than 5% we can remove above data we required approval and justification and not recomanded 

3) retain (keep as it is)

**code for the cleaning**

1) replace
    - df.loc[] = new value

2) remove
    - df.drop(index=[])
    - df.drop(columns=[]) 

3) convert
    - df['col'].astype(datatype)

4) remove duplicates 
    - df.drop_duplicates()

5) replace missing values
    - df['col'].fillna(value)
