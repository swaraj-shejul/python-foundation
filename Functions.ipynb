# User-defined Functions 
- Functions that we defined ourselves to do certain specific task are reffered as user-defined functions
- user defined functions help to decompose a large program into small segments which makes program easy to undestand, maintain and debug.
- if repeated code occurs in a program. Function can be used to include those codes and execute when needed by calling that function.
- programmars working on large project can divide the workload by making different functions.
- Syntax


def function_name(parameters):
    """
    Doc string
    """
    function body (statement)
    return [expression]

- 1 keyword def marks the start of function header
- 2 Parameters (arguments) through which we pass values to a function. these are optional
- 3 : to mark the end of function header
- 4 doc string describe what the function does. this is optional
- 5 return statement to return a value from the function. this is optional

**return**
- the return statement is used to exit a function and go back to the place from where it was called
- return statement can contain an expression which get evaluated and the value is returned.
- it there is no expression in the statement or the return statemnet itself in not present inside a function, then the function will return none object 



```python
def square_this(x):
    '''
    this function will return the squre of the given number
    '''
    return x**2
```

```python
square_this(10)
```

# Different Types of argument 

1) Positional Argument
2) Default Argument
3) keyword Argument
4) Arbitary Argument



**1) Positional Argument**
- then number of argument while defining a function and calling a funtion should be same. otherwise it throws error
- order is preserved
- its executes in the same order 


```python
def greet (name, msg):
    print ("hellow", name, msg)
```

```python
greet("swaraj","checking...")
```

**2) Default Argument**
   - while creating the function, we are assigning a default value to an argument by using the assignment operator =
   - if you won't provide value for default argument at the time of function call, it will automatically selects the default value
   - if you assign the value for default argument at the time of function call, it will consider that value which you have assigned

```python
def greet2 (name, msg="test completed..."):
    print ("hellow", name, msg)
```

```python
greet2("swaraj","test 123")
greet2("shejul")
```

**3) Keyword argument**

- at the time of calling function we will specifically declare the values for argument rather than the default order it called keyword argument 

```python
greet2(msg="keyword argument", name="keyword")
```

**4) Arbitary Arguments**
- sometimes, we do not know in advance the number of argument that will be passed into a function. Python allow us to handle this kind of situation through function call with arbitary number of arguments.

```python
def greet3(*n):
    for i in n:
        print("orbitory argument = ",i)
    
```

```python
greet3(1,2,3,4,5,6)
```
