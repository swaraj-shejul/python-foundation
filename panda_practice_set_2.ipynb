**Q18. Create a new column, age_group, categorizing customers as:
"Under 25", "25-40", and "Above 40".
Calculate the total spending for each group using the sum function.**

```python
all_data['age_group']= 'age'
```

```python
all_data.tail(1)
```

```python
for i in range(0,99457):
    if all_data.loc[i,'age'] < 25:
        all_data.loc[i,'age_group'] = "Under 25"
    elif all_data.loc[i,'age'] > 24 and all_data.loc[i,'age'] < 41:
        all_data.loc[i,'age_group'] = "25-40"
    elif all_data.loc[i,'age'] > 40:
        all_data.loc[i,'age_group'] = "Above 40"
    else:
        all_data.loc[i,'age_group'] = np.nan
```

```python
all_data
```

```python
all_data.groupby("age_group")['total_spending'].sum()
```
