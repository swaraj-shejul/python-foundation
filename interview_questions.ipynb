**Q2. difference b/w mutuable & immutable object with example?**
Mutable –
1)	It is an datatype object on which we can perform modification on the original object 
2)	The mutable data types are – list, set, dict
3)	Mutable datatypes are provide the facility to use add, remove, replace, modify the values in same original object rather than creating new.

Immutable- 
1)	It is a datatype object on which we cant perform modification in the original object 
2)	The immutable data types are int, float, bool, string, tuple 
3)	If we want to perform any modification on immutable datatype like add, replace, remove the values we always need to create new object to store the changes


**Q3. what is difference interpeter and complier?**

Interpreter -  
1)	it is an type of code execution process in which code will execute line by line. 
2)	It is very useful in troubleshooting as we will get the error at its first occurrence.
3)	It is slower in execution as compare of compiler as it is executing line by line 
Compiler –
1)	It is an type of code execution process in which complete code will executes at once and provides output
2)	it is more complex to troubleshoot as compare to interpreter because it provides all occurrence of errors and hard to focus from which point it should resolve    
3)	it is faster in execution as compare of interpreter as it executing complete code at once 


**Q4. what is difference indexing & slicing**
Indexing – 
1)	indexing is work only on such data types which supports or consists index
2)	it is use to extract single index value
3)	if the index range is not present it will through index error
4)	in indexing we need to provide single index range to extract data 
Slicing –
1)	slicing is work only on such data type which supports or consists index 
2)	it is use to extract subset of data in given index range 
3)	if the slicing range is not present it will give empty value
4)	in slicing we need to provide start:end:step range to extract subset of data
