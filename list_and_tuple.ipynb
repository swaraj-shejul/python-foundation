# List And Tuple

# List 
    - List is a mutable data type, we can modify the values of list in the original object
    - we can store n number of values of any data type in list and all are coma sepereted 
    - [] brackets are use to create the list 

# Tuple 
    - tuple is a immutable data type, we can't change the values of list in the original object
    - we can store n number of values of any data type and all are comma sceperated 
    - () are use to create the tuple

# Create

# List
# to create the empty list
        l = []
        l1 = list()
    
# to create with values 
        l = [1,2,3]
        l1 = list([1,2,3,])

# Tuple
# to create the empty list
        l = ()
        l1 = tuple()
    
# to create with values 
        l = (1,2,3)
        l1 = tuple((1,2,3))

```python
l = [1,2,3,"swaraj", True]
t= (1,2,3,"swaraj", True)

print(l,t)
```

# Nested List and Tuple
    - list inside the liat called as nested list
    - tuple inside the tuple called as nested tuple
    - we can make n number of nested list or tuple inside the list or tuple
    - Example
            l = [1,2,3,["swara",1,2], True]
            t = (1.,2,3,("swaraj",1,2), True)

```python
l = [1,2,3,["swaraj",1,2], True]
t = (1,2,3,("swaraj",1,2), True)

print(l[3][0])
print(t[3][2])
```

# Add/Insert a value

# add a single value to list

1) list.append() 
                    - is the method to add a single value at the end of the list
2) list.insert(index,object) 
                    - it will add a single value at the given index

```python
l = [1,2,3,"swaraj", True]
l.append("kumar")
print(l)
```

```python
l = [1,3,4,5,6]
l.insert(1,2)
print(l)
```

# Add multiple values in the List

# list.extend()
        - List.extend() medhod is used to insert the multiple values to list
        - get all the values in the list and give it to extend method

```python
l = [1,2,3]
l.extend([4,5,6])
print(l)
```

- concatination also one type of method to add multiple value, but it will print the output not store it in the original object

```python
l = [1,2,3]
print(l + [4,5,6])
print(l)
```

# Delete

# delete single and multiple values in the list

    * 1) list.remove()
                    - this will remove the first occurence of the value

    * 2) l.pop(index)
                    - this will remove the index value
                    - by default index is -1
                    - after deleting it will print the value of poped index
    
    * 3) l.clear()
                    - this method cleare all the values of list
            
    * 4) l.del()    
                    - this will delete the list object

```python
l = [1,2,3,"swaraj", True]
l.remove("swaraj")
print(l)

l.pop(1)
print(l)

l.clear()
print(l)
```

# some additional list attributes 

* 1) list.sort
        - use to sort the values
* 2) list.reverse()
        - use to reverse the list values 
* 3) list.copy()
        - use to make shallow copy of list 
* 4) list.count()
        - to count the number of presense/repetetion of item in the list
* 5) list.index() 
        - to check the index number of given item

```python

l = [1,6,2,3,8,9,4,5,6,7,10]
l.sort()      # it only works for yeathor numeric or str and work on only fandamental data types
print(l)

l = [1,6,2,3,[8,9,4],5,(6,7,10)]
l.reverse()
print(l)

```

```python
# sorting in descending order
l = [1,6,2,3,8,9,4,5,6,7,10]
l.sort(reverse=True)   
print(l)
```

```python
l = [1,2,3,"swaraj", True]
print(l.index("swaraj"))

l.count(1)
```
