# Dictionary 

- dict is a mutable data type which stores key:value pair in it

# creation

**empty dict**

d = {}

d = dict{}

**with value**

d = {'a':1,'b':2,'c':3}

# points 
* keys should be immutable and values can be mutable
* keys cant be duplicated values can be duplicated 
* all data types are expected in dict

# Extract 

```python
d = {'a':1,'b':2,'c':3}
d['a']
```

```python
d = {'a':1,'b':2,'c':3,'l':[1,2,3,4,5,67,9], 'd':{'a':1,'b':2}}
print(d['l'], d['d']['b'])
```

# Modification

# add

```python
d = {'a':1,'b':2,'c':3,'l':[1,2,3,4,5,67,9], 'd':{'a':1,'b':2}}
d.update({'f':"lele", 22:22})
#OR
d['n'],d['m'] = 88,99
print(d)
```

# replace

```python
d = {'a':1,'b':2,'c':3,'l':[1,2,3,4,5,67,9], 'd':{'a':1,'b':2}}
d['d'] = 'fahhhh.....'
print(d)
```

# Delete

```python
d = {'a':1,'b':2,'c':3,'l':[1,2,3,4,5,67,9], 'd':{'a':1,'b':2}}
d.pop('d', None)
print(d)
```

# attributes


```python
d = {'a':1,'b':2,'c':3,'l':[1,2,3,4,5,67,9], 'd':{'a':1,'b':2}}
d1=d.copy()
d.clear()
print(d,d1)
```

```python
print(d1.items())
print(d1.values())
print(d1.keys())
```
