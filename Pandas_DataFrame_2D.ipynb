# Extraction

```python
df = pd.DataFrame({"name":['swaraj','ujjwal','abhishek','ram'],
                  "age":[27,28,26,29],
                  "salary":[27000,27000,30000,10000]},
                 index=['a','b','c','d'])
df
```

```python
# extract 26
# 1) by using index
print(df.iloc[2, 1])

print ("=====================")
# by using index name
print(df.loc['c','age'])
```

```python
# extract last two row

# 1) by index 
print(df.iloc[2:,:])

print("=============================")
#2) by index name
print (df.loc['c':'d',:])
```

```python
# extract 2nd column

# 1) by using index

print(df.iloc[:,1])

print("=========================")

# 2) by using column name

print(df['age'])
```

```python
# extract name and salary

df[['name','salary']]
```

**By using Condition**

```python
print(df)
```

```python
# 1) age grater than 26

df[df['age'] > 26]
```

```python
# 2) extract salary equeal to 27000 and more
df[df['salary'] >= 27000]
```

```python
# 3 ) salary is 30000 and age should be 26

df[(df['salary'] == 30000) & (df['age'] == 26)]
```

```python
# age between 26 to 28

# 1) by using condition

print(df [(df['age'] >= 26) & (df['age'] <= 28) ])

print("=================================================")

# 2) by using between operator

print (df [df['age'].between(26,28)])
```

```python
# salary of 10000, 27000, 30000

df[df['salary'].isin([10000,27000,30000])]
```

# Table modification

**Combine**

- **Concat --> append :** in 2 tables, same column names with inteval values (different values)
- **Merge** --> join : in 2 tables, same column names with same values  

**Concat**

- if 2 tables have same columns name and different values then we will use concat to combine them
- we can combine verticaly or horizontaly by defining the axis

```python
pd.concat([table3,table1])
```

```python
pd.concat([table1,table3],axis=1)
```

```python
pd.concat([table1,table3],axis=0)
```

```python
pd.concat([table1,table3],axis=0,ignore_index=True)
```

**merge**

- if 2 tables have same columns names and also same values then we will use merge to combine them
- there are 4 type of joined set
- left joined - consider the left set values 
- right joined - consider the right set values
- 


```python
pd.merge(table1,table2,on="Date")
```

```python
table1 = pd.read_excel("data_import_test.xlsx",sheet_name="Sheet1")
table1
```

```python
table2 = pd.read_excel("data_import_test.xlsx",sheet_name="Sheet2")
table2
```

**Left joined**

```python
pd.merge(table1,table2, on="city",how="left")
```

**right joined**

```python
pd.merge(table1,table2, on="city",how="right")
```

**inner joint**

```python
pd.merge(table1,table2, on="city",how="inner")
```

**outer joint / full joint**

```python
pd.merge(table1,table2, on="city",how="outer")
```

```python
# how to check common values are there or not 

set1 = set(table1["city"])
print(set1)
set2 = set(table2["city"])
print(set2)
set1.intersection(set2)
```
