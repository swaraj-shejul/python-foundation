# Module

- A group of functions, variables saved to a python file (.py) which is nothing but module

**Type of Modules**

- Inbuilt Module
- Userdefined Module

**Various Possiblities of import:**

- 1 import modulename
- 2 import module1,module2, module3
- 3 import module 1 as m
- 4 import module 1 as m1, module2 as m2, module3 as m3
- 5 from module import member
- 6 from module import member1, member2, member3
- 7 from module import member1 as x
- 8 from module import*

**We can access members by using module name**
- modulename.variable
- modulename.function()

# Userdefined Modules


- to import usedefine module, the module file should be present in the working directory to import it

```python
# to check the working directory
pwd
```

```python
#other option
import os
os.getcwd()
```

- create the .py file with functions and variables and place it inside the above path to import

for example create file module_testing.py with below functions and variables 

def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

x=90
y=50

```python
import module_testing
```

```python
module_testing.sub(10,5)
```

```python
# import all the function and variable at once

from module_testing import *

print(add(1,2))
print(x,y)
```

```python
# import one by one by there name 

from module_testing import add,x

print(add(30,40))
print(x)
```

```python
# import the function from module and provide alise name to it for better calling 

import module_testing as mt

mt.mul(3,9)
```
