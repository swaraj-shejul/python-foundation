# Task 5: Statistical Operations

```python
marks = np.array([78,85,90,66,72,88,95,60])
marks
```

```python
# mean
np.mean(marks)
```

```python
# median
np.median(marks)
```

```python
# variance
np.var(marks)
```

```python
# standard deviation
np.std(marks)
```

```python
# minimum and maximum
print("minimum marks : ",np.min(marks))
print("maximum marks : ",np.max(marks))
```

```python
# Range (max - min)
np.ptp(marks)
```

# Task 7: Sales Analysis

```python
sales = np.array([1200, 1500, 900, 2000, 1800, 1700, 1600])
sales
```

```python
# 1 Total weekly sales 

print("Total weekly sales : ",np.sum(sales))
```

```python
# 2 Average daily sales
print("Avg daily sales : ",np.mean(sales))
```

```python
# Highest and Lowest sales day

print ("Highest Sale day : ",np.max(sales))
print("Lowest sales day : ",np.min(sales))
```

```python
# 4 standerd deviation of sales
print("standard deviation of sales : ",np.std(sales))
```

```python
# 5 sales above average

avg = np.mean(sales)
print("days where sales above average : ")
print("total number of days : ",np.sum(sales > avg))
print ("sales : ", np.where(sales > avg) )
```
