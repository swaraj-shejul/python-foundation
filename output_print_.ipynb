# Print

Most people think print() is simple. But for real development, it has important behavior.

Print() will print the values, variable values, booleans.

```python
# below are the usage and examples 

a = "solo leaveling"

b = 2

# chage the seperator
print (a,b, sep="-")

# .format() 
print ("the most feamos anime is running now day which name is {}, and it has {} seasons".format(a,b))

#printing variable value using f
print (f"the most feamos anime is running now day which name is {a}, and it has {b} seasons")

#new line "\n"
print (f"the most feamos anime is running now day which name is {a},\nand it has {b} seasons")
```

# Important Parameters of print()

print() has below some important parameters:

sep=
end=
file=
flust=

print(*objects, sep=' ', end='\n', file=None, flush=)

```python
# print (*objects) print will print n number of values 

print (12,166,"swaraj",2+2)
```

# sep=
by deafault print will use space as a seperator, but by using sep= function we can assign the required seperator the output


```python
a = "swaraj"
print (a,"ai", "engineer")

#sep=

print (a,"ai", "engineer", sep="-")
```

# end=

by default it is \n means new line but by using end= we can change it 

```python
a = "swaraj"
b = "shejul"
print (a)
print (b)

# now with end=

print (a, end= " ") #we changed \n means new line with space
print (b)
```

# flush=

- Normally, Python buffers output. That means: It may wait before actually printing.
- If you set: flush=True It forces immediate output.
- Why flush=True Matters?
    - In:
    - Real-time logs
    - Long-running scripts
    - AI model training progress
    - Streaming applications
    - Without flush, output may delay.

```python
print("Loading...", end="", flush=True)
```

# file=

- By default: file=None or file=sys.stdout, Meaning → print output goes to console.
- by using file= instead of console, it writes into file.
- Why This Is Important?
    - In real AI systems:
    - Logging model output
    - Saving experiment results
    - Writing training logs


```python
f = open("output.txt", "w")
print("Hello AI", file=f)
f.close()
```
