# creation and modification and Aplication of 1-D array

# creation 

**There are 4 ways to create 1-D array**

1) create with own values - np.array()
2) create with sequence values - np.arange()
3) create with random values - np.random.randint()
4) create with same values - np.zeros()

```python
# 1 create with own values - np.array()
l = [1,2,3,4,5,6, 2.4, 5.6]
la = np.array(l)
print(la)
```

```python
# 2 create with sequence values - np.arange()
a = np.arange(1,10)
print(a)
```

```python
# 3 create with random values - np.random.randint()

a = np.random.randint(1,10,5)
print(a)
```

```python
# create with same values - np.zeros()
a = np.zeros(4) # by default it will give float zero values 
print(a)
```

```python
a = np.zeros((4,),dtype=int)+20
print(a)
```

```python
a = np.ones((4,),dtype=int)
print(a)
```

```python
a = np.linspace(1,10,5)
print(a)
```

# Modify the values in Numpy

**1) Replacing the value** 

```python
l = [10,20,40,50,30,20,60,70,90]

a = np.array(l,dtype=int)
```

```python
print(a)
```

**replacing single value**

```python
# replace 50 with 20
a[3]=20
print(a)
```

**replacing multiple values**

```python
# replace last 4 values 50
a[-4::] = 50
print(a)
```

```python
# replace last 4 values with 10,20,30,40
a[5::] = [10,20,30,40]
print(a)
```

```python
#replace with list of index number [1,3,4,7,8] with 10
a[[1,3,4,7,8]]=10
print(a)
```

**base on the condition**

```python
# replace all 20 with 25
a[(a==20)] = 25
print(a)
```

**np.append function**

```python
aa = np.append(a,[15,25,50,55,"srk"])
print(aa)
```

# Delete the values from the array

**delete single value from the array**

```python
print(a)
```

```python
#delete 25 from the array
a = np.delete(a,6)
print(a)
```

**delete multiple values**

```python
#delete 2,4,5,7 index value

a = np.delete(a,[2,4,5,7])
print(a)
```

```python
#delete by condition equel to 10

a = np.delete(a,(a==10))
print(a)
```
