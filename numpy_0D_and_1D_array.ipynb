# creation and modification and Aplication of 1-D array

# creation 

**There are 4 ways to create 1-D array**

1) create with own values - np.array()
2) create with sequence values - np.arange()
3) create with random values - np.random.randint()
4) create with same values - np.zeros()

```python
# 1 create with own values - np.array()
l = [1,2,3,4,5,6, 2.4, 5.6]
la = np.array(l)
print(la)
```

```python
# 2 create with sequence values - np.arange()
a = np.arange(1,10)
print(a)
```

```python
# 3 create with random values - np.random.randint()

a = np.random.randint(1,10,5)
print(a)
```

```python
# create with same values - np.zeros()
a = np.zeros(4) # by default it will give float zero values 
print(a)
```

```python
a = np.zeros((4,),dtype=int)+20
print(a)
```
