# Mutable

        We can modify the value in the original object 

# Immutable 

    We cant modify the value in the original object



# Immutable  

        - Int
        - float
        - str
        - bool
        - tuple

# Mutable

        - List
        - set
        - Dict

# Examples

```python
# immutable

s = 10
print(id(s))

s = 20
print(id(s))
```

```python
# mutable
l = [1, 2, 3]
print(id(l))   

l.append(4)
print(id(l))
```
