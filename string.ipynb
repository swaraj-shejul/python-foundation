# String


- In python string is a sequence of charactera written inside quotes
- it can include letters,numbers, symboles, and spcases
- string are commonly used for text handling and manipulation

```python
#string
s1 = 'xyz'
s2 = "xyz"
s3 = """ this is multiline string
it can write in n numbers of lines under tripal cote """
print (s1,s2,s3)
```

- Strings are indexed sequences. Positive indices start at 0 from the left; negative indices start at -1 from the right.

# slicing

Slicing is a way to extract a portion of a string by specifying the start and end indexes. The syntax for slicing is string[start:end], 

```python
s = "swaraj"
print(s[0:6])
print(s[::-1])
print (s[3:])
print (s[:3])
```

# String Iteration

- Strings are iterable and can loop through characters one by one

```python
s = "swaraj"
for i in s:
    print (i)
```

# Strings Immutability

strings are immutable, which meanse that they cannot be changed after they created. if we need to manipulate string then there is only way to use concatenation or slicing or formatting and store it as new string

```python
s = "swaraj"
s = "S" + s[1:]
print(s)
```

# Replacing String

```python
s = "swaraj shejul"
s2 = s.replace("swaraj", "raj")
print (s2)
```

# common string methods

# len ()
- the len () function returns the total number of charecters in a string (including spaces and punctuations)

```python
s = "swaraj"
print(len(s))
```

# upper() and lower()
- upper() method converts all char to uppercase whereas lower() method converts all charaters to lowercase 


```python

print(s.upper())
print(s.lower())
```

# concatenating and repeating string
- a string can be combined by using + operator
- string can be reapeat multiple times by using * operator


```python
print (s + " " + s)
print (s * 3)
```

# Formating String
- preferred way to formate string is by using f-strings and the another way is by using format() method

```python
name = "swaraj"
age = 26
print (f"Name: {name}, age:{age}")

s = "My name is {} and i am year old.".format("swaraj",26)
print (s)
```

# String Membership Testing
- in keyword checks if a particular substring is present in a string.
    

```python
s = "jhsfashgjhrguhsdjfhvj"
print("fas" in s)
print("xxx" in s)
```
