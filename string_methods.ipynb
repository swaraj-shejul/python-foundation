# Data Finding Methods

# 1) str.index()
    - writes index value for given substring if it's present or if not it throws error 

```python
s = "swaraj"
s.index("r")
```

```python
# it prints the index of first occurrence only
s.index("a")
```

```python
# if the substring is not found it raises an error  
s.index("x")
```

# 2) str.find()
    - writes index value for given substring if it's present or if not it will give -1 as output

```python
s = "swaraj"
s.find("r")
```

```python
# it prints the index of first occurrence only
s.find("a")
```

```python
# if the substring is not found it raises an error  
s.find("x")
```

# Data filteretion methods

# 1) str.strip
    - use to remove the starting and ending spaces of string

```python
s = "       swaraj pravin shejul         "
s.strip()
```

# 2) str.lstrip()
    - removes left side starting space 

```python
s = "    swaraj pravin shejul         "
s.lstrip()
```

# 3) str.rstrip()

```python
s = "        swaraj pravin shejul         "
s.rstrip()
```

# Replacing the string value

# 1) str.replace()
    - replaces the substring with given substring

```python
s = "swaraj kumar"
s1 = s.replace("kumar", "shejul")
print(s1)
```

# 2) str.split()
    - split the string and store in list
    - by default it takes space as an delimeter to split the list
    - we can add the delimeter by which it should split

```python
s = "swaraj pravin shejul"
s1 = s.split()
print(s1)
print(type(s1))
```

```python
s = "swaraj-pravin-shejul-        "
s1 = s.split("-")
print(s1)
```
