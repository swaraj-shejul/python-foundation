# Create

**Pandas series can be created using List**

```python
l = [10,20,30]
s = pd.Series(l)
print(s)
```

```python
type(s)
```

```python
s.shape
```

```python
s.ndim
```

**We can provide names to the index**

```python
l = [10,20,30]
s = pd.Series(l,index=['a','b','c'])
print(s)
```

**Create 1D Series using 1D Array**

```python
a = np.array([10,20,30])
s = pd.Series(a,index=['a','b','c'])
print(s)
```

**change the index names**

```python
s = s.loc[['c','b','a']]
print(s)
```

```python
a = np.random.randint(1,100,10)
s = pd.Series(a)
print(s)
```

```python
s = pd.Series(np.ones((5,),dtype=int)+5)
print(s)
```

**Create Series using Dictionary**

```python
d = {'a':10,'b':20,'c':30}
s = pd.Series(d)
print(s)
```

**In Series we can add any data type no restriction like array to only have combination of 0D**

```python
l = [1,2,3,'a',[1,2,3],{4,5,6},{'a':1,'b':2}]
s = pd.Series(l)
print(s)
```

# Extract

```python
l = [15,19,23,24,37,42,46]
s = pd.Series(l,index=['a','b','c','d','e','f','g'])
print(s)
```

**extract 19,37,46**

**1st way by using index (iloc)**

```python
s.iloc[[1,4,-1]]
```

**2nd way by using index name (loc)**

```python
s.loc[['b','e','g']]
```

```python
# extract from 19 to 42
s.loc['b':'f']
```

**By condition**

```python
# extract values grate than 15

s[s>15]
```

```python
# extract values not grate than 15

s[~(s>30)]
```

# Modification

**Replace**

```python
# replace 15,19,23 with 25,29,33

s.loc[['a','b','c']] = [25,29,33]
print(s)
```

**Add new value**

```python
# add 50 to h

s['h'] = 50
print(s)
```

```python
# add 50,51 to i j

s['i'],s['j'] = 50,51
print(s)
```

```python
print(s)
```

**Delete**

**Remove single vlaue**

```python
# remove d value 

s = s.drop(index=['d'])
print(s)
```

**remove muliple values**

```python
# remove c and e

s = s.drop(index=['c','e'])
print(s)
```

- if index name is not present it will through key error

```python
s+s2
```

Why all values are NaN
s has indices: a, b, f, g
s2 has indices: x, y, z, o

There are no common indices, so Pandas tries to align them and finds nothing to match.

So internally it's like:

a: 15 + NaN → NaN
x: NaN + 1  → NaN

That’s why everything becomes NaN.
